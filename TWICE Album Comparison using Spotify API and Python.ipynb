## Standard Deviation and Means

Standard deviation tells us about the variability of values in a data set. It is a measure of dispersion, showing how spread out the data points are around the mean. Together with the mean, standard deviation can also indicate percentiles for a normally distributed population (jdmeducational.com).

<b>A large value for standard deviation</b> means that the data is spread far out, with some of it far away from the mean.  That is, on average, a given data point is far from the mean.

<b>A small value for standard deviation</b> means that the data is clustered near the mean.  That is, on average, a given data point is close to the mean.

<b>A zero value for standard deviation</b> means that all of the data has the same value (which is also the value of the mean).  This situation is rare, but it is possible

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
# prepare lists
features = ['acousticness', 'danceability', 'energy', 'key', 'loudness', 'speechiness', 'tempo', 'valence']
list_album=[]
# standard deviation list for each features
s_acousticness=[]
s_danceability=[]
s_energy=[]
s_key=[]
s_loudness=[]
s_speechiness=[]
s_tempo=[]
s_valence=[]
# mean list for each features
m_acousticness=[]
m_danceability=[]
m_energy=[]
m_key=[]
m_loudness=[]
m_speechiness=[]
m_tempo=[]
m_valence=[]
```

```python
def getstats(df):
    # get album name
    album = df['album'].head(1)
    # standard deviation
    s = df[features].std()
    sd=pd.concat([album, s])
    # mean value
    m = pd.concat([album, df[features].mean(skipna=False)])
    # mean value of standard deviation
    sdm = [album[0], s.mean()]
    
    # create featurelist
    featurelist(sd,m)
    
    return sd, m, sdm

def featurelist(sd,m):
    if sd[0] in list_album:
        return
    list_album.append(sd[0])
    
    s_acousticness.append(sd['acousticness'])
    s_danceability.append(sd['danceability'])
    s_energy.append(sd['energy'])
    s_key.append(sd['key'])
    s_loudness.append(sd['loudness'])
    s_speechiness.append(sd['speechiness'])
    s_tempo.append(sd['tempo'])
    s_valence.append(sd['valence'])
    
    m_acousticness.append(m['acousticness'])
    m_danceability.append(m['danceability'])
    m_energy.append(m['energy'])
    m_key.append(m['key'])
    m_loudness.append(m['loudness'])
    m_speechiness.append(m['speechiness'])
    m_tempo.append(m['tempo'])
    m_valence.append(m['valence'])
```

```python
# 2015 <the story begins>
tsb = pd.read_csv('tsb.csv')
sd_tsb, m_tsb, sdm_tsb = getstats(tsb)
```

## Danceability, Energy, and Loudness

These three features are related with each other. In general, high danceability also have high energy and usually high valence(although not always but majority).

<b>Danceability:</b> Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.

<b>Acousticness:</b> A measure from 0.0 to 1.0 of whether the track is acoustic.
Energy: Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy.

<b>Loudness:</b> The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track. Values typical range between -60 and 0 db.

# Valence and Tempo

Valence and tempo dictates the mood of a song. In this section, we'll take a look which album is the moodiest.

<b>Valence:</b> A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).

<b>Tempo:</b> The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.

Valence of the albums seems to be distributed evenly throughout the upperhalf of the values. All of TWICEs albums give off positive vibes in general. Fan-favourite albums seems to be on the higher valence average as well. ONCEs are such sucker for feel-good music.

## Speechiness

<b>Speechiness:</b> Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.

Since all of the tracks analyzed are music track, we are expecting average values of each album to be less than 0.33

```python
plt.figure(figsize=(15,10))
plt.barh(i, m_speechiness, width, label='mean', color='darkgray')
#plt.barh(i+width, m_speechiness, width, label='mean', color='lightgray')
plt.axvline(x=0.33, color='k',linestyle='--', linewidth=1) 
plt.yticks(i, list_album)
plt.ylabel('Albums')
plt.xlabel('Standard Deviation / Mean')
plt.legend(loc='best')
```
